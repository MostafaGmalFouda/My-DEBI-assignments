```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = r"D:\Gemy Study\Course_Data Science\Session_21\train.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
dnull = df.isnull().sum()
dnull
```

```python
dnull[dnull > 0].plot(kind='bar',color='r')
plt.title('Missing Values in the Dataset')
plt.xlabel("Columns")
plt.ylabel("Number of Missing Values")
plt.show()
```

```python
sex_count = df['Sex'].value_counts()
sex_count
```

```python
sex_count.plot(kind='bar',color=['black','red'])
plt.title('Passengers in each')
plt.ylabel("Number of Passengers")
plt.show()
```

```python
sex_count.plot(kind='pie',autopct='%1.1f%%',colors=['purple','red'])
plt.title('Passengers in each')
plt.legend(["Male","Female"])
plt.show()
```

```python
df['Age'].hist(color='b')
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
age_null = df['Age'].isnull().sum()
print(age_null)
```

```python
df[df['Survived']==0]['Age'].hist()
```

```python
Pclass_count = df['Pclass'].value_counts()
Pclass_count = Pclass_count.sort_index()
Pclass_count
```

```python
Pclass_count.plot(kind='bar', color=['black','red','blue'])
plt.title('Passengers in each')
plt.ylabel("Number of Passengers")
plt.show()
```

```python
Pclass_count.plot(kind='pie',autopct='%1.1f%%',colors=['purple','red','blue'])
plt.title('Passengers in each')
plt.legend(["1st Class","2nd Class","3rd Class"])
```

```python
Parch_Survived_count = df[df['Survived']==1]['Parch'].value_counts()
Parch_Survived_count
```

```python
count_1 = pd.crosstab(df['Parch'],df['Survived'])
count_1.plot(kind='bar', stacked=False, color=['r','b'])
plt.title('Survived by Parch')
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python
Sibsp_Survived_count = df[df['Survived']==1]['SibSp'].value_counts()
Sibsp_Survived_count
```

```python
count_2 = pd.crosstab(df['SibSp'],df['Survived'])
count_2.plot(kind='bar', stacked=False, color=['black','r'])
plt.title('Survived by SibSp')
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python
Pclass_Survived_count = df[df['Survived']==1]['Pclass'].value_counts()
Pclass_Survived_count = Pclass_Survived_count.sort_index()
Pclass_Survived_count
```
